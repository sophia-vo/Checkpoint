```python
import pandas as pd
```

```python
data = pd.read_csv('data/keystroke_data_combined.csv')
data.head()
```

```python
data.dtypes
```

```python
data[~data['Parkinsons'].apply(lambda x: isinstance(x, bool))]
```

```python
data = data.dropna(subset='Parkinsons')
data['Parkinsons'] = data['Parkinsons'].astype(bool)
```

```python
data.dtypes
```

```python
parkinsons = data[data['Parkinsons']]
parkinsons
```

```python
parkinsons = parkinsons.dropna(subset=['BirthYear', 'Gender', 'Parkinsons', 'DiagnosisYear'])
parkinsons
```

```python
parkinsons['DiagnosisYear'] = parkinsons['DiagnosisYear'].astype(int)
parkinsons['BirthYear'] = parkinsons['BirthYear'].astype(int)
parkinsons['Gender'] = parkinsons['Gender'].astype(str)
parkinsons['current_year'] = parkinsons['Date'] // 10000 + 2000
```

```python
averages = parkinsons.groupby(['UserKey', 'BirthYear', 'Gender', 'Parkinsons', 'DiagnosisYear', 'current_year', 'Direction'])[['LatencyTime']].median()
```

```python
averages.loc['5USOYSDCXB']
```

```python
for user in averages.reset_index().UserKey.unique():
    if averages.loc[user].shape[0] == 9:
        print(user, averages.loc[user].reset_index().set_index('Direction')['LatencyTime'].to_dict())
```

```python
i = 1
for user in averages.reset_index().UserKey.unique():
    user_df = averages.loc[user].reset_index()
    if user_df.shape[0] == 9:
        user_age = user_df['current_year'].max() - user_df['BirthYear'].max()
        user_gender = user_df['Gender'].max()
        print(f'<option> value="{user}">Person {i}, {user_age}{user_gender[0]}</option>')
        i += 1
```
